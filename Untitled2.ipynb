```python
from google.colab import files
uploaded = files.upload()


import fitz
from docx import Document
from sentence_transformers import SentenceTransformer, util
from tabulate import tabulate
import matplotlib.pyplot as plt


model = SentenceTransformer('all-MiniLM-L6-v2')

def extract_text(file_name):
    if file_name.endswith(".pdf"):
        doc = fitz.open(file_name)

        text = ""
        for page in doc:
            text += page.get_text()
        return text

    elif file_name.endswith(".docx"):
        doc = Document(file_name)
        text = ""
        for para in doc.paragraphs:
            text += para.text
        return text

    return ""

job_description = input("📄 Enter Job Description: ")


def extract_skills_from_jd(jd):
    words = jd.lower().replace(",", "").split()
    return list(set(words))

jd_skills = extract_skills_from_jd(job_description)


def extract_skills(text, jd_skills):
    text = text.lower()
    return [skill for skill in jd_skills if skill in text]


def missing_skills(jd_skills, resume_skills):
    return list(set(jd_skills) - set(resume_skills))


def explain_result(skills_found, missing, score):
    if len(skills_found) == 0:
        return "No relevant skills found."

    explanation = f"Matched: {', '.join(skills_found)}. "
    explanation += f"Missing: {', '.join(missing)}. "

    if score > 0.75:
        explanation += "Strong candidate."
    elif score > 0.5:
        explanation += "Moderate candidate."
    else:
        explanation += "Weak candidate."

    return explanation


results = []

for file_name in uploaded.keys():
    print(f"\nProcessing: {file_name}")

    resume_text = extract_text(file_name)

    if resume_text.strip() == "":
        print("No text found, skipping...")
        continue

    skills_found = extract_skills(resume_text, jd_skills)
    missing = missing_skills(jd_skills, skills_found)

    emb1 = model.encode(resume_text, convert_to_tensor=True)
    emb2 = model.encode(job_description, convert_to_tensor=True)

    similarity = float(util.cos_sim(emb1, emb2))

    skill_score = len(skills_found) / max(len(jd_skills), 1)

    final_score = (0.7 * similarity) + (0.3 * skill_score)

    explanation = explain_result(skills_found, missing, final_score)

    results.append((file_name, final_score, skills_found, missing, explanation))


results.sort(key=lambda x: x[1], reverse=True)


table_data = []

for item in results:
    name, score, skills_found, missing, explanation = item

    if score > 0.75:
        status = "🔥 Strong"
    elif score > 0.5:
        status = "⚡ Moderate"
    else:
        status = "❌ Weak"

    table_data.append([
        name,
        round(score, 2),
        status,
        ", ".join(skills_found),
        ", ".join(missing)
    ])

print("\n🏆 RESUME RANKING TABLE\n")

print(tabulate(
    table_data,
    headers=["Resume", "Score", "Level", "Matched Skills", "Missing Skills"],
    tablefmt="fancy_grid"
))

if len(results) > 0:
    top = results[0]
    print("\n🔥 TOP CANDIDATE 🔥")
    print(f"Name: {top[0]}")
    print(f"Score: {top[1]:.2f}")
    print(f"Matched Skills: {top[2]}")
    print(f"Missing Skills: {top[3]}")
    print(f"Explanation: {top[4]}")


names = [item[0] for item in results]
scores = [item[1] for item in results]

plt.figure()
plt.bar(names, scores)
plt.xlabel("Resumes")
plt.ylabel("Score")
plt.title("Resume Ranking Visualization")
plt.xticks(rotation=30)
plt.show()
```
